```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
print(os.getcwd())

# Importing the dataset
# Replace 'your_dataset_url' with the actual URL or file path
url = "forest+fires/forestfires.csv"
df = pd.read_csv(url)

# Exploratory Data Analysis (EDA)
# Explore the dataset's structure and summary statistics
display(df.head(15))


# Documentation and Reporting
# Document the entire analysis process, including code and explanations
# Summarize key findings in a clear and concise report
```

```python
# Visualize key features using plots
numeric_cols = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Value')
    plt.show()
```

```python
# Hypothesis Formulation
# Formulate at least two hypotheses related to the dataset

# Hypothesis 1: The month in the year also affects the amount of rainfall.
# Hypothesis 2: If the rainfall ("rain") decreases, then the burned area of the forest ("area") will increase.
```

```python
# Hypothesis Testing
# Test the formulated hypotheses using statistical tests or visualizations

# Hypothesis 1: The month in the year also affects the amount of rainfall.
# Grouping the data by month and calculating the average rainfall
avg_rainfall_month = df.groupby('month')['rain'].mean().reset_index()

# Ordering the months 
months_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# Plotting the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='month', y='rain', data=avg_rainfall_month, order=months_order, ci=None)
plt.title('Average Rainfall by Month')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm/m2)')
plt.show()

```

```python
# Hypothesis Testing
# Hypothesis 2: If the rainfall ("rain") decreases, then the burned area of the forest ("area") will increase.

# Scatter plot of rain vs area:
plt.figure(figsize=(8, 6))
plt.scatter(df['rain'], df['area'], alpha=0.5)
plt.title('Rainfall vs Burned Area')
plt.xlabel('Rainfall (mm/m2)')
plt.ylabel('Burned Area (ha)')
plt.show()
```

```python
# Conclusion: From the graphs plotted, we can clearly see that there's a strong correlation in both cases.
# For hypothesis 1, it stated that the month in the year also affects the amount of rainfall. We can see that
# in the month of August, that's when the majority of the rainfall in mm/m2 with some scattering rainfall in 
# the month of March and July. In the remaining months, there was none or too little to be significant enough. 
#
# For hypthesis, 2, it stated that iff the rainfall ("rain") decreases, then the burned area of the forest ("area") 
# will increase. This statement is also provened by the plot. From the scattered plot, there are more burned area (ha)
# when the rain was none or minimal (0-1 mm/m2).
#
# Thus, from the information that have been gathered, it can be inferred that in months without rain, there is 
# more burned area or it is more likely to have forest fires.
```
